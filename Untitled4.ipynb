```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
hotels = pd.read_csv("https://osf.io/y6jvb/download")
```

```python
hotels
```

```python
hotels.shape
```

```python
# Filter by accomadation type
hotels = hotels[hotels["accommodation_type"] == "Hotel"]
# Filter by city actual
hotels = hotels[hotels["city_actual"] == "Vienna"]
#Filter by stars between 3 and 4
hotels = hotels[(hotels["stars"] >= 3) & (hotels["stars"] <= 4)]
# Filter out null stars
hotels = hotels[hotels["stars"].notnull ()]
# Filter by price <= 600
hotels = hotels [hotels["price"]<= 600]
```

```python
hotels.filter(["price", "distance"]).describe(percentiles=[0.25, 0.5, 0.75, 0.95]).T
```

```python
sns.scatterplot(data=hotels, x="distance", y="price", color="purple", alpha=0.5,s=12)
#set x and y limits
plt.xlim(0, 7)
plt.ylim(0, 400)
plt.xticks(range(0, 8))
plt.yticks(np.arange(0, 401, 50))
# Set axis labels
plt.xlabel("Distance to city center (miles)")
plt.ylabel("Price (US dollars)")
# Show the plot
plt.show()
```

```python
hotels["dist2"] = np.where(hotels["distance"] >= 2, "Far", "Close")
hotels["Eprice_cat2"] = hotels.groupby("dist2")["price"].transform("mean")
```

```python
import statsmodels.formula.api as smf
```

```python
simple_reg = smf.ols("price ~ distance", data=hotels).fit()
print(simple_reg.summary())
```

```python
hetero_rob_reg = smf.ols("price ~ distance", data=hotels).fit(cov_type ="HC3")
print(hetero_rob_reg.summary())
```

```python
from stargazer.stargazer import Stargazer
```

```python
table = Stargazer([simple_reg, hetero_rob_reg])
table.rename_covariates({"Intercept": "Constant"})
table.custom_columns(["Homoskedastic SE", "Heteroskedastic robust SE"], [1, 1])
table
```

```python
html = table.render_html()
with open("my_regression_table.html", "w") as f:
    f.write(html)
#open('lin_reg.html','w').write(table.render_html())
```

```python
hotels["predprice"] = simple_reg.fittedvalues
hotels["e"] = simple_reg.resid
# e = residual value
```

```python
hotels["lnprice"] = np.log(hotels["price"])
```

```python
hotels["distance2"] = np.where(hotels["distance"] < 0.05, 0.05,hotels["distance"])
```

```python
hotels["lndistance"] = np.log(hotels["distance2"])
```

```python
hotels.filter(["price", "lnprice"]).describe(percentiles=[0.25, 0.5, 0.75, 0.95]).T
```

```python
reg1 = smf.ols("price ~ distance", data=hotels).fit()
print(reg1.summary())
```

```python
sns.lmplot(x="distance", y="price", data=hotels, scatter_kws={"color": "red","alpha": 0.5, "s": 20}, line_kws={"color": "blue"}, ci=None)
plt.show()
```

```python
reg2 = smf.ols("price ~ lndistance", data=hotels).fit()
print(reg2.summary())
```

```python
sns.lmplot(x="lndistance", y="price", data=hotels, scatter_kws={"color": "red","alpha": 0.5, "s": 12})
plt.xlim(-3, 2)
plt.ylim(0, 400)
plt.xlabel("ln(distance to city center, miles)")
plt.ylabel("Price (US dollars)")
plt.show()
```

```python
reg3 = smf.ols("lnprice ~ distance", data=hotels).fit()
print(reg3.summary())
```

```python
reg4 = smf.ols("lnprice ~ lndistance", data=hotels).fit()
print(reg4.summary())
```

```python
table = Stargazer([reg1, reg2, reg3, reg4])
table.rename_covariates({"Intercept": "Constant"})
table.custom_columns(["Level-Level", "Level-Log", "Log-Level", "Log-Log"], [1,1, 1, 1])
table
```

```python
hotels["dist_sq"] = hotels["distance"]**2
hotels["dist_cb"] = hotels["distance"]**3
```

```python
reg5 = smf.ols("price ~ distance + dist_sq", data=hotels).fit()
print(reg5.summary())
```

```python
sns.lmplot(x="distance", y="price", data=hotels, order=2, ci=68,scatter_kws={"color": "red", "alpha": 0.6}, line_kws={"color": "blue"})
plt.xlabel("Distance to city center (miles)")
plt.ylabel("Price (US dollars)")
plt.show()
```

```python
reg6 = smf.ols("price ~ distance + dist_sq + dist_cb", data=hotels).fit()
print(reg6.summary())
```

```python
sns.lmplot(x="distance", y="price", data=hotels, ci=68, order=3,scatter_kws={"color": "blue"})
plt.show()
```

```python
table = Stargazer([reg1, reg5, reg6])
table.rename_covariates({"Intercept": "Constant", "dist_cb":" Distance Cube","dist_sq":"Distance Squared"})
table.custom_columns(["Linear", "Squared", "Cubic"], [1, 1, 1])
table
```
